## Odometer created using pandas Series
The odometer on the bicycle was recorded at the end of each day as below.

```python
import pandas as pd
import numpy as np

odometer = pd.Series([55, 120, 182, 240, 299, 
                      360, 410, 465, 523, 580, 
                      645, 702, 768, 830, 895, 
                      965, 1023, 1086, 1151, 1215, 
                      1281, 1345, 1407, 1470, 1519])

print(odometer)
```

## Miles ridden each day calculated

```python
daily_miles = {}
previous_miles = 0
print("Daily miles ridden")

for day, odometer_miles in odometer.items():
    current_miles = odometer_miles - previous_miles
    daily_miles[day] = current_miles
    print(current_miles)
    previous_miles = odometer_miles
```
